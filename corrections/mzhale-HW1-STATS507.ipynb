```python
#-----------------------------------------------------------------------------
#--------------------------------Question 1-----------------------------------
#-----------------------------------------------------------------------------

#-----------------------------Part A------------------------------------------

fib_history={}
def fib_rec(n,a,b):
    """
    Generates the n-th element of the Fibonnaci sequence by a recursive method

    Parameters
    ----------
    n : integer
        Index of the Fibonnaci sequence.
    a : integer
        First element of the Fibonnaci sequence.
    b : integer
        Second element of the Fibonnaci sequence.

    Returns
    -------
    integer
        n-th element of the Fibonnaci sequence.

    """
    if n in fib_history:
        return fib_history[n]
    if n == 0:
        return a
    elif n == 1:
        return b
    elif n >= 2:
        fib_history[n] = fib_rec(n-1,a,b)+fib_rec(n-2,a,b)
        return fib_rec(n-1,a,b) + fib_rec(n-2,a,b)
    
    
# Test
#print(fib_rec(7,0,1),fib_rec(11,0,1),fib_rec(13,0,1))

#-----------------------------Part B------------------------------------------

def fib_for(n):
    """
    Generates the n-th element of the Fibonnaci sequence by a for-loop method

    Parameters
    ----------
    n : integer
        Index of the Fibonnaci sequence.


    Returns
    -------
    integer
        n-th element of the Fibonnaci sequence.
        
    """
    fib_seq = [0,1]
    if n in [0,1]:
        return fib_seq[n]
    elif n >= 2:
        for i in range(2,n+1):
            fib_seq.append(fib_seq[i-1] + fib_seq[i-2])
        return fib_seq[n]
# Test
#print(fib_for(7),fib_for(11),fib_for(13))
        
#-----------------------------Part C------------------------------------------

def fib_whl(n):
    """
    Generates the n-th element of the Fibonnaci sequence by a while method

    Parameters
    ----------
    n : integer
        Index of the Fibonnaci sequence.


    Returns
    -------
    integer
        n-th element of the Fibonnaci sequence.
        
    """
    fib_seq = [0,1]
    if n in [0,1]:
        return fib_seq[n]
    elif n >= 2:
        index = 2
        while index <= n:
            fib_seq.append(fib_seq[index-1] + fib_seq[index-2])
            index  += 1
        return fib_seq[n]
# Test
#print(fib_whl(7),fib_whl(11),fib_whl(13))
        
#-----------------------------Part D------------------------------------------
        
phi = (1 + np.sqrt(5))/2
def fib_rnd(n):
    """
    Generates the n-th element of the Fibonnaci sequence by the 
    rounding method

    Parameters
    ----------
    n : integer
        Index of the Fibonnaci sequence.


    Returns
    -------
    integer
        n-th element of the Fibonnaci sequence.
        
    """
    return np.round((phi**n) / np.sqrt(5))
# Test
#print(fib_rnd(7),fib_rnd(11),fib_rnd(13))   
    
#-----------------------------Part E------------------------------------------

def fib_flr(n):
    """
    Generates the n-th element of the Fibonnaci sequence by the 
    truncation method

    Parameters
    ----------
    n : integer
        Index of the Fibonnaci sequence.


    Returns
    -------
    integer
        n-th element of the Fibonnaci sequence.
        
    """
    return np.floor((phi**n)/np.sqrt(5)+0.5)
# Test
#print(fib_flr(7),fib_flr(11),fib_flr(13)) 

#-----------------------------Part F------------------------------------------

#Time of fib_rec
time_rec, time_for, time_whl, time_rnd, time_flr = [], [], [], [], []
for i in range(1,5000):
    start = time.time()
    fib_rec(i,0,1)
    end = time.time()
    time_rec.append(end-start)
    
    start = time.time()
    fib_for(i)
    end = time.time()
    time_for.append(end-start)
    
    start = time.time()
    fib_whl(i)
    end = time.time()
    time_whl.append(end-start)
    
    start = time.time()
    fib_rnd(i)
    end = time.time()
    time_rnd.append(end-start)
    
    start = time.time()
    fib_flr(i)
    end = time.time()
    time_flr.append(end-start)

df = pd.DataFrame ({'Recursive':[np.median(time_rec)], 
    'For':[np.median(time_for)],'While':[np.median(time_whl)], 
    'Round':[np.median(time_rnd)], 'Floor':[np.median(time_flr)]})

display(HTML(df.to_html(index=False)))
```
